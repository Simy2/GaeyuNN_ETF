```python
import pandas as pd
import chardet

# 파일 경로
file_path = 'data/표준코드.csv'

# 파일의 인코딩 감지
with open(file_path, 'rb') as f:
    result = chardet.detect(f.read())

encoding = result['encoding']
print(encoding)

# 감지된 인코딩으로 파일 읽기
try:
    df = pd.read_csv(file_path, encoding=encoding)
    print(df.head())
except UnicodeDecodeError:
    print(f"파일을 {encoding} 인코딩으로 읽는 데 실패했습니다.")
```
